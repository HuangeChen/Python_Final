# Part 1 Yearly Hiring and Total Hiring on Each State

```python
import pandas as pd
import numpy as np
import datetime

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_H1B=pd.read_csv("/Users/huangechen/Desktop/final/data/h1b_Petitions.csv")
```

```python
df_H1B.head()
```

```python
df_H1B['STATE']=df_H1B['WORKSITE'].str.split(',').str[1]
```

```python
state=df_H1B.groupby(['STATE', 'YEAR']).size()
```

```python
state_year=state.unstack()
```

```python
state_year['TOTAL'] = state_year.sum(axis=1)
```

```python
state_year
```

```python
state_year.to_csv('/Users/huangechen/Desktop/final/output/ANA4_Part1_State.csv')
```
